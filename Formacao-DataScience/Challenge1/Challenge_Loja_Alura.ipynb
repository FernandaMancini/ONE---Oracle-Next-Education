```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

# **Faturamento Total de Cada Loja**

```python
faturamento1 = sum(loja["Preço"])
faturamento2 = sum(loja2["Preço"])
faturamento3 = sum(loja3["Preço"])
faturamento4 = sum(loja4["Preço"])

print(f"O faturamento total da loja 1 é de R${faturamento1:.2f}")
print(f"O faturamento total da loja 2 é de R${faturamento2:.2f}")
print(f"O faturamento total da loja 3 é de R${faturamento3:.2f}")
print(f"O faturamento total da loja 4 é de R${faturamento4:.2f}")
```

# **Gráfico de Colunas Faturamento**

```python
faturamento_lojas = {
    'Loja 1': loja['Preço'].sum(),
    'Loja 2': loja2['Preço'].sum(),
    'Loja 3': loja3['Preço'].sum(),
    'Loja 4': loja4['Preço'].sum()
}

plt.figure(figsize=(8,5))
plt.bar(faturamento_lojas.keys(), faturamento_lojas.values(), color=['#4CAF50', '#2196F3', '#FF9800', '#9C27B0'])
plt.title('Faturamento Total por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y')
plt.show()
```

# **Gráfico de Linha Faturamento por Categoria**

```python
df_categorias = pd.concat(
    [produtos1, produtos2, produtos3, produtos4],
    axis=1
)
df_categorias.columns = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
df_categorias = df_categorias.fillna(0)

plt.figure(figsize=(12,6))
for loja_nome in df_categorias.columns:
    plt.plot(df_categorias.index, df_categorias[loja_nome], marker='o', label=loja_nome)

plt.title("Faturamento por Categoria de Produto em Cada Loja")
plt.xlabel("Categoria do Produto")
plt.ylabel("Faturamento (R$)")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Gráfico Boxplot Distribuição de Preços**

```python
plt.figure(figsize=(10,6))

plt.boxplot([loja["Preço"], loja2["Preço"], loja3["Preço"], loja4["Preço"]],
            labels=["Loja 1", "Loja 2", "Loja 3", "Loja 4"])

plt.title("Distribuição de Preços por Loja")
plt.ylabel("Preço (R$)")
plt.grid(True)
plt.show()
```

# **Gráfico de Dispersão Relação Preço e Avaliação**

```python
plt.figure(figsize=(10,6))

plt.scatter(loja["Preço"], loja["Avaliação da compra"], alpha=0.5, label='Loja 1')
plt.scatter(loja2["Preço"], loja2["Avaliação da compra"], alpha=0.5, label='Loja 2')
plt.scatter(loja3["Preço"], loja3["Avaliação da compra"], alpha=0.5, label='Loja 3')
plt.scatter(loja4["Preço"], loja4["Avaliação da compra"], alpha=0.5, label='Loja 4')

plt.title("Relação entre Preço e Avaliação da Compra")
plt.xlabel("Preço (R$)")
plt.ylabel("Avaliação (0 a 5)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_regionais = pd.DataFrame({
    'Loja 1': loja['Local da compra'].value_counts(),
    'Loja 2': loja2['Local da compra'].value_counts(),
    'Loja 3': loja3['Local da compra'].value_counts(),
    'Loja 4': loja4['Local da compra'].value_counts()
}).fillna(0).astype(int)

df_regionais.plot(kind='bar', figsize=(15, 7), title='Comparativo Regional entre Lojas')
plt.ylabel('Número de Compras')
plt.xlabel('Local da compra')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# **Conclusão**

```python
criterios = [
    "Faturamento Total",
    "Venda Ctg. Mais Vendida",
    "Avaliação Média",
    "Prod. Mais Vendido",
    "Ticket Médio",
    "Frete Médio"
]

lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]

dados = [
    ["R$ 1,534,509.12", 465, 3.98, "Micro-ondas: 60", "R$ 34.69", "R$ 650.49"],  # Loja 1
    ["R$ 1,488,459.06", 442, 4.04, "Init. em prog.: 65", "R$ 33.62", "R$ 630.97"],  # Loja 2
    ["R$ 1,464,025.03", 499, 4.05, "Kit banquetas: 57", "R$ 33.07", "R$ 620.61"],  # Loja 3
    ["R$ 1,384,497.58", 480, 4.00, "Cama box: 62", "R$ 31.28", " R$587.15"]   # Loja 4
]

fig, ax = plt.subplots(figsize=(15, 4))
ax.axis('off')

tabela = ax.table(
    cellText=dados,
    rowLabels=lojas,
    colLabels=criterios,
    cellLoc='center',
    loc='center'
)
tabela.auto_set_font_size(False)
tabela.set_fontsize(10)
tabela.scale(1.2, 1.5)
plt.title("Comparativo de Desempenho por Loja", fontsize=14, weight='bold')
plt.show()
```
